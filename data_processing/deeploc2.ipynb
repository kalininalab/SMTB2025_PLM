```python
import pandas as pd
```

```python
splits = {'train': 'deeploc-train.parquet', 'validation': 'deploc-val.parquet', 'test': 'deeploc-test.parquet'}
df = pd.read_parquet("hf://datasets/bloyal/deeploc/" + splits["train"])
df2 = pd.read_parquet("hf://datasets/bloyal/deeploc/" + splits["test"])
df3 = pd.read_parquet("hf://datasets/bloyal/deeploc/" + splits["validation"])
```

11 classes info

```python
df4 = pd.DataFrame({})
df4["Sequence"]=df["Sequence"]
df4["Cytoplasm"]=df["Cytoplasm"]
df4["Nucleus"]=df["Nucleus"]
df4["Extracellular"]=df["Extracellular"]
df4["Mitochondrion"]=df["Mitochondrion"]
df4["Cell membrane"]=df["Cell membrane"]
df4["Plastid"]=df["Plastid"]
df4["Endoplasmic reticulum"]=df["Endoplasmic reticulum"]
df4["Lysosome/Vacuole"]=df["Lysosome/Vacuole"]
df4["Golgi apparatus"]=df["Golgi apparatus"]
df4["Peroxisome"]=df["Peroxisome"]
df4["split"]="train"

df5 = pd.DataFrame({})
df5["Sequence"]=df2["Sequence"]
df5["Cytoplasm"]=df2["Cytoplasm"]
df5["Nucleus"]=df2["Nucleus"]
df5["Extracellular"]=df2["Extracellular"]
df5["Mitochondrion"]=df2["Mitochondrion"]
df5["Cell membrane"]=df2["Cell membrane"]
df5["Plastid"]=df2["Plastid"]
df5["Endoplasmic reticulum"]=df2["Endoplasmic reticulum"]
df5["Lysosome/Vacuole"]=df2["Lysosome/Vacuole"]
df5["Golgi apparatus"]=df2["Golgi apparatus"]
df5["Peroxisome"]=df2["Peroxisome"]
df5["split"]="test"

df6 = pd.DataFrame({})
df6["Sequence"]=df3["Sequence"]
df6["Cytoplasm"]=df3["Cytoplasm"]
df6["Nucleus"]=df3["Nucleus"]
df6["Extracellular"]=df3["Extracellular"]
df6["Mitochondrion"]=df3["Mitochondrion"]
df6["Cell membrane"]=df3["Cell membrane"]
df6["Plastid"]=df3["Plastid"]
df6["Endoplasmic reticulum"]=df3["Endoplasmic reticulum"]
df6["Lysosome/Vacuole"]=df3["Lysosome/Vacuole"]
df6["Golgi apparatus"]=df3["Golgi apparatus"]
df6["Peroxisome"]=df3["Peroxisome"]
df6["split"]="valid"
```

```python
dff = pd.concat([df4, df5, df6], ignore_index=True)
```

```python
ind = []
for i in range(len(dff)):
    ind.append(f"P{i:05d}")
dff["ID"] = ind
```

```python
dff
```

```python
dff.to_csv('deeploc2_data.csv', index = False)
```
