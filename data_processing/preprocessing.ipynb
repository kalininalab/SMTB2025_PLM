```python
import pandas as pd
import numpy as np
```

```python
splits = {'train': 'fluorescence_train.csv', 'validation': 'fluorescence_valid.csv', 'test': 'fluorescence_test.csv'}
df = pd.DataFrame()
for split in splits:
    df2 = pd.read_csv("hf://datasets/proteinea/fluorescence/" + splits[split])
    df2["split"] = split
    if split == "validation":
        df2["split"] = "valid"
    df = pd.concat([df, df2], ignore_index = True)
```

```python
df
```

```python
df2 = pd.DataFrame(
    {
        'ID': np.arange(len(df)),
        "Sequence": df["primary"],
        "label": df["log_fluorescence"],
        "split": df["split"]
    }
)

df2
```
