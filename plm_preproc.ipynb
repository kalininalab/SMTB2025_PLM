importing the data from the hugging face about the stability of proteins

```python
import pandas as pd
```

```python

splits = {'train': 'data/train-00000-of-00001.parquet', 'valid': 'data/valid-00000-of-00001.parquet', 'test': 'data/test-00000-of-00001.parquet'}
df = pd.read_parquet("hf://datasets/proteinglm/stability_prediction/" + splits["train"])
df3 = pd.read_parquet("hf://datasets/proteinglm/stability_prediction/" + splits["test"])
df4 = pd.read_parquet("hf://datasets/proteinglm/stability_prediction/" + splits["valid"])
```

```python
df2 = pd.DataFrame({})
```

```python
df2["Sequence"]=df["seq"]
df2["label"]=df["label"]
df2["split"]="train"
```

```python
df5 = pd.DataFrame({})
```

```python
df5["Sequence"]=df3["seq"]
df5["label"]=df3["label"]
df5["split"]="test"
```

```python
df6 = pd.DataFrame({})
```

```python
df6["Sequence"]=df4["seq"]
df6["label"]=df4["label"]
df6["split"]="valid"
```

```python
dff = pd.concat([df2, df5, df6], ignore_index=True)
```

```python
dff
```

```python
ind = []
for i in range(len(dff)):
    ind.append(f"p{i:05d}")
ind
```

```python
dff["ID"] = ind
```

```python
dff.to_csv('datagy.csv', index = False)
```
